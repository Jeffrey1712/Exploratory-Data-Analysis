### HOUSE MARKET

```python
import pandas as pd
```

```python
superficie = pd.read_csv("src/house_market/superficie.csv",header = None)
sdb = pd.read_csv("src/house_market/nombre_de_sdb.csv",header = None)
chambre = pd.read_csv("src/house_market/nombre_de_chambres.csv",header = None)
price = pd.read_csv("src/house_market/house_price.csv",index_col=0)
```

```python
superficie.head()
```

```python
sdb.head()
```

```python
chambre.head()
```

```python
price.head()
```

```python
superficie.columns = ['id', 'surface']
sdb.columns = ['id','bathrooms']
chambre.columns = ['id', 'rooms']
```

```python
superficie.head()
```

```python
new_df = superficie.merge(sdb, on="id").merge(chambre, on="id").merge(price, on="id")
new_df
```

```python
new_df["surface"].mean().__round__(2)
```

```python
print("The average number of room is {}, and the median number is {}"
      .format(new_df["rooms"].mean(), new_df['bathrooms'].median()))
```

```python
new_df.describe()
```

```python
new_df.groupby('rooms')['house_price'].mean()
```

```python
new_df["house_size"] = new_df['surface'].apply(lambda x : "very large" if x > 25000
                                           else "large" if (x > 20000) and (x <=25000)
                                           else "medium" if (x > 15000) and (x <=20000)
                                           else "small" if (x > 10000) and (x<= 15000)
                                           else "very small")

new_df
```

```python
new_df.groupby("house_size")["house_price"].mean()
```
