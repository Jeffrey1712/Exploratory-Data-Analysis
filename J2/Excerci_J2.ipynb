### Heatmaps ðŸ”¥

```python
dataset = pd.read_csv("src/house_pricing2.csv")
dataset = dataset.iloc[:,-15:]
dataset.head()
```

```python
dataset.describe(include='all')
```

```python
dataset.describe() # pour garder que les colonnes numÃ©riques
```

```python
new_cols = ["WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch", "PoolArea", "MiscVal", "MoSold", "YrSold", "SalePrice"]
dataset = dataset[new_cols]
dataset
```

```python
dataset.info()
```

```python
correlation = dataset.corr()
correlation 
```

```python
sns.heatmap(correlation, annot=True)
```

```python
abs(correlation['SalePrice']).sort_values(ascending = False)
```
